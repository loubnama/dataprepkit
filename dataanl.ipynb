```python
import pandas as pd
import numpy as np
def get_data_path():
    file_path = input("enter the path ")
    return file_path

def read_data(file_path):
    if file_path.endswith('.csv'):
        df = pd.read_csv(file_path)
    elif file_path.endswith('.xlsx') or file_path.endswith('.xls'):
        df = pd.read_excel(file_path)
    elif file_path.endswith('.json'):
        df = pd.read_json(file_path)
    else:
        raise ValueError(" الملف غير مدعوم. CSV، Excel، أو JSON.")
    return df

file_path = get_data_path()
df = read_data(file_path)
df
```

```python

print(df.shape)
print(df.dtypes)
df.head()
```

```python
print("Data summary:")
def calculate_max_and_mean_and_mode(df):
    max_values = df.max(numeric_only=True)
    mean_values = df.mean(numeric_only=True)
    str_col= df.select_dtypes(include=['object']).columns
    mode_values = df[str_col].mode()
    return max_values,mean_values,mode_values

max_values, mean_values,mode_values= calculate_max_and_mean_and_mode(df)
print("max")
print(max_values)
print("mean")
print(mean_values)
print("mode")
print(mode_values)
```

```python
df.isnull().mean()
```

```python
print("Delete columns that contain many missing values")
threshold =0.5
print(threshold)
columns_to_drop = df.columns[df.isnull().mean()>threshold]
df = df.drop(columns=columns_to_drop)
df
```

```python
print("Category Data Encoding:")
from sklearn.preprocessing import MinMaxScaler
numeric_columns = df.select_dtypes(include=['int', 'float']).columns
num_data = df[numeric_columns]
m=MinMaxScaler()
numeric_columns_tr=m.fit_transform(num_data)
df[numeric_columns] = numeric_columns_tr
df
```

```python
print("Categorical Data Encoding:")
from sklearn.preprocessing import LabelEncoder
binary_columns = [col for col in df.columns if (df[col].nunique()== 2 or df[col].nunique()< 2) and df[col].dtype == 'object']
le = LabelEncoder()
for col in binary_columns:
    df[col] = le.fit_transform(df[col])


```

```python
print("Handling Missing Values:")
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer

numeric_columns = df.select_dtypes(include=['int', 'float']).columns
categorical_columns = df.select_dtypes(include=['object']).columns

numeric_imputer = SimpleImputer(strategy='mean')

categorical_imputer = SimpleImputer(strategy='most_frequent')
df[numeric_columns] = numeric_imputer.fit_transform(df[numeric_columns])
df[categorical_columns] = categorical_imputer.fit_transform(df[categorical_columns])
df


```
